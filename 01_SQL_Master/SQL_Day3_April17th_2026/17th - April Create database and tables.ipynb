Here are 20 essential Databricks SQL commands for daily data engineering:

Database/Schema Management:

SHOW DATABASES - List all databases
USE database_name - Switch to a database
CREATE DATABASE db_name - Create new database
DROP DATABASE db_name - Delete database
DESCRIBE DATABASE db_name - Show database details
Table Operations:

SHOW TABLES - List tables in current database
DESCRIBE TABLE table_name - Show table schema
SHOW CREATE TABLE table_name - View table DDL
DROP TABLE table_name - Delete table
TRUNCATE TABLE table_name - Remove all rows
Data Manipulation:

SELECT * FROM table_name LIMIT 10 - Preview data
INSERT INTO table_name VALUES (...) - Add rows
UPDATE table_name SET col = value WHERE ... - Modify rows
DELETE FROM table_name WHERE ... - Remove rows
MERGE INTO target USING source ON ... - Upsert data
Metadata & Monitoring:

DESCRIBE HISTORY table_name - Show table version history
SHOW COLUMNS IN table_name - List columns
SHOW PARTITIONS table_name - View partitions
OPTIMIZE table_name - Compact small files
VACUUM table_name - Remove old files
These commands cover database exploration, table management, data operations, and Delta Lake optimization—core daily tasks in data engineering.
