```sql
-- ============================================================
--  UNILEVER SQL MASTER — DATABRICKS
--  STEP 1: Create Database
--  STEP 2: Create all Managed Delta Tables
--  STEP 3: Insert Data (20 rows each)
--  STEP 4: Verify
--  Author : Paresh Ranjan Rout
-- ============================================================


-- ============================================================
-- STEP 1 — Create Database
-- ============================================================

CREATE DATABASE IF NOT EXISTS unilever_db
COMMENT 'Unilever SQL Practice — Paresh Ranjan Rout';

USE unilever_db;


-- ============================================================
-- STEP 2A — Create Table: customers
-- ============================================================

CREATE TABLE IF NOT EXISTS customers (
  customer_id   INT     COMMENT 'Primary Key',
  customer_name STRING  COMMENT 'Full name',
  city          STRING  COMMENT 'City of customer',
  region        STRING  COMMENT 'North / South / East / West / Central',
  segment       STRING  COMMENT 'Retail / Wholesale / Distributor',
  email         STRING  COMMENT 'Email — some rows have NULL intentionally',
  joined_date   DATE    COMMENT 'Customer onboarding date'
)
USING DELTA
COMMENT 'Unilever Customer Master Table';


-- ============================================================
-- STEP 2B — Create Table: products
-- ============================================================

CREATE TABLE IF NOT EXISTS products (
  product_id   INT     COMMENT 'Primary Key',
  product_name STRING  COMMENT 'Product name with size/variant',
  category     STRING  COMMENT 'Personal Care / Home Care / Foods',
  brand        STRING  COMMENT 'Brand name (Dove, Surf Excel, Knorr...)',
  unit_price   DOUBLE  COMMENT 'Price per unit in INR'
)
USING DELTA
COMMENT 'Unilever Product Catalogue';


-- ============================================================
-- STEP 2C — Create Table: orders
-- ============================================================

CREATE TABLE IF NOT EXISTS orders (
  order_id    INT     COMMENT 'Primary Key',
  customer_id INT     COMMENT 'Foreign Key -> customers.customer_id',
  product_id  INT     COMMENT 'Foreign Key -> products.product_id',
  order_date  DATE    COMMENT 'Date order was placed',
  quantity    INT     COMMENT 'Number of units ordered',
  amount      DOUBLE  COMMENT 'Total order value in INR',
  status      STRING  COMMENT 'Delivered / Shipped / Cancelled',
  region      STRING  COMMENT 'Region where order was placed'
)
USING DELTA
COMMENT 'Unilever Sales Orders';


-- ============================================================
-- STEP 2D — Create Table: employees
-- ============================================================

CREATE TABLE IF NOT EXISTS employees (
  employee_id INT     COMMENT 'Primary Key',
  name        STRING  COMMENT 'Full name',
  department  STRING  COMMENT 'Sales / Marketing / Finance / HR / Supply Chain',
  region      STRING  COMMENT 'Region of posting',
  salary      DOUBLE  COMMENT 'Annual salary in INR',
  manager_id  INT     COMMENT 'FK -> employees.employee_id (NULL for top level)',
  join_date   DATE    COMMENT 'Date of joining Unilever'
)
USING DELTA
COMMENT 'Unilever Employee Master — includes manager hierarchy for SELF JOIN practice';


-- ============================================================
-- STEP 2E — Create Table: monthly_sales
-- ============================================================

CREATE TABLE IF NOT EXISTS monthly_sales (
  year_month  STRING  COMMENT 'Format YYYY-MM e.g. 2026-01',
  region      STRING  COMMENT 'Region name',
  category    STRING  COMMENT 'Product category',
  total_sales DOUBLE  COMMENT 'Total revenue for that month + region + category',
  order_count INT     COMMENT 'Number of orders placed'
)
USING DELTA
COMMENT 'Monthly Aggregated Sales — used for Window Function practice';


-- ============================================================
-- STEP 3A — Insert Data: customers (20 rows)
-- ============================================================

INSERT INTO customers VALUES
(1,  'Ravi Sharma',    'Mumbai',    'West',    'Retail',      'ravi@gmail.com',    '2021-03-10'),
(2,  'Priya Nair',     'Bangalore', 'South',   'Wholesale',   'priya@yahoo.com',   '2020-06-15'),
(3,  'Amit Verma',     'Delhi',     'North',   'Retail',       NULL,               '2019-11-20'),
(4,  'Sunita Rao',     'Hyderabad', 'South',   'Wholesale',   'sunita@hul.com',    '2022-01-05'),
(5,  'Karan Singh',    'Pune',      'West',    'Retail',      'karan@gmail.com',   '2021-08-22'),
(6,  'Meena Joshi',    'Chennai',   'South',   'Distributor', 'meena@hul.com',     '2020-12-01'),
(7,  'Deepak Gupta',   'Kolkata',   'East',    'Retail',      'deepak@yahoo.com',  '2023-02-14'),
(8,  'Anita Patel',    'Ahmedabad', 'West',    'Wholesale',    NULL,               '2022-07-30'),
(9,  'Vijay Kumar',    'Jaipur',    'North',   'Retail',      'vijay@gmail.com',   '2021-05-18'),
(10, 'Neha Mishra',    'Lucknow',   'North',   'Distributor', 'neha@hul.com',      '2020-03-25'),
(11, 'Rajesh Pillai',  'Kochi',     'South',   'Retail',      'rajesh@gmail.com',  '2019-09-12'),
(12, 'Kavya Reddy',    'Bangalore', 'South',   'Wholesale',   'kavya@hul.com',     '2023-04-07'),
(13, 'Suresh Patil',   'Pune',      'West',    'Retail',      'suresh@yahoo.com',  '2022-11-15'),
(14, 'Pooja Iyer',     'Chennai',   'South',   'Distributor',  NULL,               '2021-07-03'),
(15, 'Arjun Mehta',    'Delhi',     'North',   'Wholesale',   'arjun@gmail.com',   '2020-10-28'),
(16, 'Divya Krishnan', 'Mumbai',    'West',    'Retail',      'divya@hul.com',     '2023-01-19'),
(17, 'Sanjay Tiwari',  'Bhopal',    'Central', 'Retail',      'sanjay@gmail.com',  '2022-05-06'),
(18, 'Rekha Sharma',   'Jaipur',    'North',   'Wholesale',   'rekha@yahoo.com',   '2021-12-11'),
(19, 'Mohan Das',      'Kolkata',   'East',    'Distributor', 'mohan@hul.com',     '2020-08-20'),
(20, 'Lakshmi Venkat', 'Hyderabad', 'South',   'Retail',      'lakshmi@gmail.com', '2023-03-14');


-- ============================================================
-- STEP 3B — Insert Data: products (15 rows)
-- ============================================================

INSERT INTO products VALUES
(101, 'Dove Soap 100g',          'Personal Care', 'Dove',        45.00),
(102, 'Surf Excel 1kg',          'Home Care',     'Surf Excel',  120.00),
(103, 'Lux Shampoo 180ml',       'Personal Care', 'Lux',         95.00),
(104, 'Knorr Soup Mix 50g',      'Foods',         'Knorr',       55.00),
(105, 'Lifebuoy Handwash 250ml', 'Personal Care', 'Lifebuoy',    75.00),
(106, 'Rin Detergent 500g',      'Home Care',     'Rin',         60.00),
(107, 'Ponds Face Cream 50g',    'Personal Care', 'Ponds',       110.00),
(108, 'Kissan Jam 500g',         'Foods',         'Kissan',      130.00),
(109, 'Domex Toilet Cleaner',    'Home Care',     'Domex',       85.00),
(110, 'TRESemme Conditioner',    'Personal Care', 'TRESemme',    199.00),
(111, 'Hellmanns Mayonnaise',    'Foods',         'Hellmanns',   170.00),
(112, 'Comfort Fabric Softener', 'Home Care',     'Comfort',     145.00),
(113, 'Closeup Toothpaste 80g',  'Personal Care', 'Closeup',     65.00),
(114, 'Horlicks 200g',           'Foods',         'Horlicks',    180.00),
(115, 'Brooke Bond Tea 250g',    'Foods',         'Brooke Bond', 90.00);


-- ============================================================
-- STEP 3C — Insert Data: orders (20 rows)
-- ============================================================

INSERT INTO orders VALUES
(1001,  1,  101, '2026-01-05',  10,  4500,  'Delivered', 'West'),
(1002,  2,  102, '2026-01-12',  20, 24000,  'Delivered', 'South'),
(1003,  3,  103, '2026-01-20',   5,  4750,  'Shipped',   'North'),
(1004,  4,  104, '2026-02-03',  15,  8250,  'Delivered', 'South'),
(1005,  5,  105, '2026-02-15',   8,  6000,  'Delivered', 'West'),
(1006,  6,  106, '2026-02-22',  25, 15000,  'Cancelled', 'South'),
(1007,  7,  107, '2026-03-01',  12, 13200,  'Delivered', 'East'),
(1008,  8,  108, '2026-03-10',  30, 39000,  'Delivered', 'West'),
(1009,  9,  109, '2026-03-18',   6,  5100,  'Shipped',   'North'),
(1010, 10,  110, '2026-03-25',  10, 19900,  'Delivered', 'North'),
(1011, 11,  111, '2026-04-02',   8, 13600,  'Delivered', 'South'),
(1012, 12,  112, '2026-04-08',  15, 21750,  'Delivered', 'South'),
(1013, 13,  113, '2026-04-10',  20, 13000,  'Shipped',   'West'),
(1014, 14,  114, '2026-01-28',  12, 21600,  'Delivered', 'South'),
(1015, 15,  115, '2026-02-10',  18, 16200,  'Delivered', 'North'),
(1016, 16,  101, '2026-02-18',   5,  2250,  'Cancelled', 'West'),
(1017, 17,  102, '2026-03-05',  10, 12000,  'Delivered', 'Central'),
(1018, 18,  103, '2026-03-20',   7,  6650,  'Delivered', 'North'),
(1019, 19,  104, '2026-04-01',  20, 11000,  'Delivered', 'East'),
(1020, 20,  105, '2026-04-05',   9,  6750,  'Shipped',   'South');


-- ============================================================
-- STEP 3D — Insert Data: employees (20 rows)
-- ============================================================

INSERT INTO employees VALUES
(201, 'Rahul Khanna',     'Sales',        'North',   95000,  NULL, '2018-04-01'),
(202, 'Sneha Agarwal',    'Sales',        'South',   88000,   201, '2019-07-15'),
(203, 'Tarun Bose',       'Marketing',    'East',    72000,   201, '2020-01-10'),
(204, 'Ritu Pandey',      'Finance',      'North',   65000,   201, '2021-03-20'),
(205, 'Alok Srivastava',  'Sales',        'West',    91000,   202, '2019-11-05'),
(206, 'Preethi Nambiar',  'Marketing',    'South',   70000,   202, '2020-06-22'),
(207, 'Harsh Vardhan',    'Supply Chain', 'North',   78000,   203, '2021-08-15'),
(208, 'Manasi Deshpande', 'Finance',      'West',    67000,   203, '2022-02-01'),
(209, 'Kunal Jain',       'Sales',        'North',   85000,   204, '2022-05-10'),
(210, 'Anjali Menon',     'HR',           'South',   60000,   204, '2020-09-30'),
(211, 'Vivek Tripathi',   'Supply Chain', 'Central', 75000,   205, '2021-12-20'),
(212, 'Nandini Iyer',     'Marketing',    'South',   69000,   205, '2023-01-14'),
(213, 'Abhishek Roy',     'Sales',        'East',    82000,   206, '2022-07-07'),
(214, 'Shruti Malhotra',  'Finance',      'North',   64000,   206, '2023-03-25'),
(215, 'Girish Patil',     'HR',           'West',    58000,   207, '2021-10-11'),
(216, 'Puja Saha',        'Supply Chain', 'East',    76000,   207, '2022-08-19'),
(217, 'Rohit Bansal',     'Sales',        'North',   90000,   208, '2020-04-14'),
(218, 'Kavita Chopra',    'Marketing',    'West',    71000,   208, '2021-06-28'),
(219, 'Nitin Desai',      'Finance',      'West',    63000,   209, '2023-06-01'),
(220, 'Sonal Gupta',      'HR',           'North',   57000,   209, '2023-09-05');


-- ============================================================
-- STEP 3E — Insert Data: monthly_sales (20 rows)
-- ============================================================

INSERT INTO monthly_sales VALUES
('2025-10', 'West',  'Personal Care', 320000, 42),
('2025-10', 'South', 'Home Care',     210000, 35),
('2025-10', 'North', 'Foods',         185000, 28),
('2025-10', 'East',  'Personal Care', 145000, 22),
('2025-11', 'West',  'Personal Care', 345000, 46),
('2025-11', 'South', 'Home Care',     230000, 38),
('2025-11', 'North', 'Foods',         195000, 30),
('2025-11', 'East',  'Personal Care', 160000, 25),
('2025-12', 'West',  'Personal Care', 390000, 52),
('2025-12', 'South', 'Home Care',     275000, 44),
('2025-12', 'North', 'Foods',         220000, 34),
('2025-12', 'East',  'Personal Care', 180000, 28),
('2026-01', 'West',  'Personal Care', 310000, 40),
('2026-01', 'South', 'Home Care',     205000, 33),
('2026-01', 'North', 'Foods',         175000, 27),
('2026-01', 'East',  'Personal Care', 135000, 21),
('2026-02', 'West',  'Personal Care', 355000, 48),
('2026-02', 'South', 'Home Care',     240000, 39),
('2026-02', 'North', 'Foods',         200000, 31),
('2026-02', 'East',  'Personal Care', 165000, 26);


-- ============================================================
-- STEP 4 — Verify Everything
-- ============================================================

-- Check all tables exist
SHOW TABLES IN unilever_db;

-- Check row counts
SELECT 'customers'    AS table_name, COUNT(*) AS rows FROM customers    UNION ALL
SELECT 'products',                   COUNT(*)          FROM products      UNION ALL
SELECT 'orders',                     COUNT(*)          FROM orders        UNION ALL
SELECT 'employees',                  COUNT(*)          FROM employees     UNION ALL
SELECT 'monthly_sales',              COUNT(*)          FROM monthly_sales;

-- Preview each table
SELECT * FROM customers    LIMIT 5;
SELECT * FROM products     LIMIT 5;
SELECT * FROM orders       LIMIT 5;
SELECT * FROM employees    LIMIT 5;
SELECT * FROM monthly_sales LIMIT 5;

-- Confirm Delta format
DESCRIBE TABLE EXTENDED customers;

-- ============================================================
-- ✅ ALL DONE — Tables are ready for Day 1 practice!
-- ============================================================
```
